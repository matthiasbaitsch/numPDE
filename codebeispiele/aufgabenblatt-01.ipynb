## Aufgabe 2

```julia
l = 10
EA = 1
C = 2
S = 5
n = 1
F = 10

x = parameter(0 .. l)
u = 10 - x / 20
u = x

fplot(u, u)
```

```julia
a(u, u) = EA * integrate(u' * u', 0 .. l) + C * integrate(u * u, 0 .. l) + S * u(l) * u(l)
b(u) = n * integrate(u, 0 .. l) + F * u(0)

println(a(u, u))
println(b(u))
```

```julia
a(u, u)
```

### Aufgabe 3

```julia
l = 3.7
EI = 2.2
k = 3.9
q = 1.1

x = parameter(0 .. l)
w = (12 * EI * l * q + k * l^3 * q * x - 2 * k * l * q * x^3 + k * q * x^4) / (24 * EI * k)
w = 1 + x

fplot(w, w, zero(x))
```

```julia
a(w, w) = EI * integrate(w'' * w'', 0 .. l) + k * w(0) * w(0) + k * w(l) * w(l)
b(w) = q * integrate(w, 0 .. l)

w = sin(x) + x
println(a(w, w))
println(b(w))
```

```julia
fplot(EI * w'''' - q)
```

```julia
println(k * w(0) + EI * w'''(0))
println(k * w(l) - EI * w'''(l))
println(w''(0))
println(w''(l))
```
